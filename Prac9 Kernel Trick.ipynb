```python
from sklearn.datasets import make_circles
```

```python
x ,y = make_circles(n_samples=100 ,factor=.1 ,noise=.1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[: ,0] ,x[:,1] ,c=y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain ,xtest ,ytrain ,ytest = train_test_split(x ,y ,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import accuracy_score
```

```python
classifier = SVC(kernel='linear')
classifier.fit(xtrain ,ytrain)
pred =classifier.predict(xtest)
accuracy_score(ytest ,pred)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x ,y ,clf = classifier)
```

```python
classifier2 = SVC(kernel = 'rbf')
classifier2.fit(xtrain ,ytrain)
pred =classifier2.predict(xtest)
accuracy_score(ytest ,pred)
```

```python
plot_decision_regions(x ,y ,clf = classifier2)
```

```python
classifier3 = SVC(kernel='poly' ,degree=2)
classifier3.fit(xtrain ,ytrain)
pred = classifier3.predict(xtest)
accuracy_score(ytest ,pred)
```

```python
plot_decision_regions(x ,y ,clf =classifier3)
```
